```python
#9. Create summary report with below column by combination of Gender, Serious_delinquency.
# a. Number of customers
# b. Percentage of customers
# c. Average Revolving utilization
# d. Standard deviation with in the utilization
# e. Average monthly income
# f. Maximum Monthly income
# g. Standard deviation with in the income
# h. Average Age
# i. Average Dependents
```

```python
grpd = pd.merge(cust_data,cust_demo, how = 'inner').groupby(['Gender','SeriousDlqin2yrs'])
```

```python
No_of_customers = pd.DataFrame(grpd.count().ID)
```

```python
Percentage_of_customers = pd.DataFrame(grpd.count().ID.apply(lambda x: x/pd.merge(cust_data,cust_demo, how = 'inner').ID.count()))
```

```python
Average_Revolving_Utilization = pd.DataFrame(grpd.mean().RevolvingUtilization)
```

```python
STD_Utilization = pd.DataFrame(grpd.std().RevolvingUtilization)
```

```python
Average_Monthly_Income = pd.DataFrame(grpd.mean().MonthlyIncome)
```

```python
Maximum_Monthly_Income = pd.DataFrame(grpd.max().MonthlyIncome)
```

```python
STD_Income = pd.DataFrame(grpd.std().MonthlyIncome)
```

```python
Average_Age = pd.DataFrame(grpd.mean().age)
```

```python
Average_Dependents = pd.DataFrame(grpd.mean().No_of_dependents)
```

```python
column_names = ['Number_of_customers','Percentage_of_customers','Average_Revolving_Utilization','STD_Utilization','Average_Monthly_Income','Maximum_Monthly_Income','STD_Income','Average_Age','Average_Dependents']
```

```python
indexes = ['Gender0,Deliquency0','Gender0,Deliquency1','Gender1,Deliquency0','Gender1,Deliquency1']
```

```python
pd.DataFrame(pd.concat([Number_of_customers,Percentage_of_customers,Average_Revolving_Utilization,STD_Utilization,Average_Monthly_Income,Maximum_Monthly_Income,STD_Income,Average_Age,Average_Dependents],axis = 1).values,columns = column_names,index = indexes)
```
