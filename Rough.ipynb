```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("X Vs Y Chart")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.xlabel('X Axis')
plt.ylabel("Y Axis")
plt.title("1st Chart")
plt.subplot(1,2,2)
plt.plot(y,x, 'g*-')
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("2nd Chart")
plt.tight_layout()
plt.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([1,0.5,0.5,1])
axes.plot(x,y,'b')
axes.set_xlabel("X Axis")
axes.set_ylabel("Y Axis")
axes.set_title("Chart")
```
