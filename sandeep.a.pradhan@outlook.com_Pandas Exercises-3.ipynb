### Step 3. Assign it to a variable called chipo.

```python
chipo = pd.read_table('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv')
```

### Step 4. How many products cost more than $10.00?

```python
import re
sum(chipo['item_price'].replace('\$','',regex = True).apply(pd.to_numeric) > 10.00)
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
chipo[chipo[['item_name','item_price']].duplicated() == False][['item_name','item_price']]
```

### Step 6. Sort by the name of the item

```python
chipo[chipo[['item_name','item_price']].duplicated() == False][['item_name','item_price']].sort_values(by = 'item_name')
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
print (chipo[chipo['item_price'].replace('\$','',regex = True).apply(pd.to_numeric) == chipo['item_price'].replace('\$','',regex = True).apply(pd.to_numeric).max()].quantity)
```

### Step 8. How many times were a Veggie Salad Bowl ordered?

```python
sum(chipo['item_name'] == 'Veggie Salad Bowl')
```

### Step 9. How many times people orderd more than one Canned Soda?

```python
sum(chipo[chipo['item_name'] == 'Canned Soda'].quantity >1)
```
