### Step 9. Get the last business day of each month

```python
dates = pd.date_range('1980-12-12', '2014-07-08', freq='BM')
LBM = []
for x in dates:
    for y in apple.Date:
        if x == y:
            LBM.append(y)
print(pd.DataFrame(LBM).apply(pd.to_datetime))
```
